On peut utiliser la librairie Polars pour faire du SQL, et à l'inverse, apprendre du SQL avec Polars...

Date : 07-03-23

```python
import polars as pl
```

```python
# Récupération avec l'instruction scan_scv()
# pour pouvoir travailler sur le fichier
# au format SQL
df = pl.scan_csv('data/titanic.csv')
```

```python
# Traitements pour travailler sous SQL
sql = pl.SQLContext()
sql.register('df', df)
```

```python
# Filtre opéré sous SQL : 
# 5 premières lignes des personnes âgées de + de 30 ans
sql.query("""
          SELECT Survived, Name, Sex, Age
          FROM df WHERE Age > 30 LIMIT 5  
          """)
```
